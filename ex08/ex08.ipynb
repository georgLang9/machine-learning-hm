```python
from sklearn.decomposition._pca import PCA
from sklearn.datasets import load_iris
import pandas as pd
from matplotlib import pyplot as plt


iris_dataset = load_iris()
X, y = iris_dataset['data'], iris_dataset['target']

model = PCA(n_components=2)
model = model.fit(X)
```

```python
model_components = model.components_
model_var = model.explained_variance_

model_transform = model.transform(X)

plt.scatter(model_transform[:, 0], model_transform[:, 1], c=y)
```

```python
from sklearn.cluster import KMeans
cluster = KMeans(n_clusters=3)
cluster = KMeans.fit(cluster, model_transform)
cluster_predict = KMeans.predict(cluster, model_transform)

plt.scatter(model_transform[:, 0], model_transform[:, 1], c=cluster.fit_predict(model_transform))
plt.show()

# Cluster predict scatter plot
plt.scatter(model_transform[:, 0], model_transform[:, 1], c=cluster_predict)
plt.show() 

# Plot the transformed features with colors representing the predicted clusters
plt.scatter(model_transform[:, 0], model_transform[:, 1], c=cluster_predict, cmap='viridis', edgecolor='k')
plt.title('KMeans Clusters')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.show()

# Plot the transformed features with colors representing the actual species
plt.scatter(model_transform[:, 0], model_transform[:, 1], c=y, cmap='viridis', edgecolor='k')
plt.title('Actual Species')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.show()
```
