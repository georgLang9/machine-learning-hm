```python
import pandas
import math
import matplotlib.pyplot

ice = pandas.read_csv('IceCream.csv')
temp = ice['Temperature']
sold = ice['SoldIceCream']
ice_plt = matplotlib.pyplot.scatter(temp, sold)
```

```python
import pandas

def train(X, Y, steps, eta):
    w0, w1 = 0, 0
    for i in range(steps):
        dw0, dw1, = 0, 0
        for j in range(len(X)):
            x = X[j]
            y = Y[j]
            dw0 = dw0 - y + w1 * x + w0 
            dw1 = dw1 - x * y + w1 * x * x + w0 * x
        w0 = w0 - eta * dw0
        w1 = w1 - eta * dw1
    return w0, w1


w0, w1 = train(temp, sold, 200000, 0.0001) 
print(w0, w1)
```

```python
def predict(x, w0, w1):
    sold = w1 * x + w0
    return sold

sold_pred = predict(25, w0, w1)
print(sold_pred)
```

```python
x_min = min(temp)
x_max = max(temp)

y_min = predict(x_min, w0, w1)
y_max = predict(x_max, w0, w1)

x = [x_min, x_max]
y = [y_min, y_max]

matplotlib.pyplot.scatter(temp, sold)
matplotlib.pyplot.plot(x, y)
matplotlib.pyplot.show()
```
